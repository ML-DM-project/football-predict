```python
df = pd.read_csv("match_all_statistic.csv", usecols=list_column)

df.head()
```

```python
df
```

```python
y1 = df.iloc[:, 85]
y2 = df.iloc[:, 83]
y = pd.concat([y1, y2], axis=1)
y
```

```python
X = df.drop(["id", "awayScoreCurrent", "homeScoreCurrent"], axis=1)
X
```

```python
# Create label based on goal results
# create a list of our conditions
conditions = [
    (y['homeScoreCurrent'] < y['awayScoreCurrent']),
    (y['homeScoreCurrent'] == y['awayScoreCurrent']),
    (y['homeScoreCurrent'] > y['awayScoreCurrent'])
]

values = ["Lose", "Draw", "Win"]

y = pd.DataFrame(np.select(conditions, values))

y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train
```

```python
X_train = X_train.fillna(X_train.mean())
X_test = X_test.fillna(X_train.mean())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train[0]
```
